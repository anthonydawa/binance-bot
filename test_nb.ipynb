```python
import pandas as pd
```

```python
df = pd.read_csv('backtest_data_isolated\BTCUSDT-5 May 2013-17 June 2021.csv')
df
```

```python
df_15m = df.loc[df['timeframe'] == '15m']
```

```python
from kline_initial import klines_initial

def get_unix_days(d):
    latest_unix_days = klines_initial('BTCUSDT', '1d', d+5)[-d-1:]
    latest = [ x[0] for x in latest_unix_days ]
    return latest
get_unix_days(3)
```

```python
def win_intervals(d):
    # use pd and get all pd conditions 
    da = get_unix_days(d)
    results = []
    for idx,_ in enumerate(range(d)):
        # tdf.append(df.loc[(da[idx] <= df['d']) & ( df['d'] <= da[idx+1])])
        ivl = df_15m.loc[(1528002900000 <= df_15m['d']) & ( df_15m['d'] <= 1564012800000)]
        grp = ivl.groupby(['method']).sum().sort_values('result', ascending=False)
        results.append(grp)

    for result in results:  
        f = []
        for idx, row in result.iterrows():
            f.append({idx:row['result']})
        print(f)

win_intervals(3)
```
